Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis

PIMA Diabetes Dataset

```python
diabetes_dataset=pd.read_csv('/content/diabetes.csv')
```

```python
diabetes_dataset.head()
```

```python
diabetes_dataset.shape
diabetes_dataset.shape?
```

```python
diabetes_dataset.describe()
```

```python
diabetes_dataset['Outcome'].value_counts()
```

0-Non_diabetic

1-Diabetic

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
X=diabetes_dataset.drop(columns='Outcome',axis=1)
Y=diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

Data Standardization

```python
scaler=StandardScaler?
```

```python
scaler=StandardScaler()
```

```python
standardized_data=scaler.fit(X).transform(X)
```

```python
print(standardized_data)
```

```python
X=standardized_data
print(X)
```

Train_Test Split

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

```python
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

Model Evaluation

```python
X_train_prediction=classifier.predict(X_train)
Training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print('Accuracy of the training data:',Training_data_accuracy)
```

```python
X_test_prediction=classifier.predict(X_test)
Testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('Accuracy of the test data:',Testing_data_accuracy)
```

Predictive System

```python
input_data = (5,166,72,19,175,25.8,0.587,51)

# Change the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshape the numpy array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# Get the column names from the original dataset (excluding 'Outcome')
feature_names = diabetes_dataset.drop(columns='Outcome', axis=1).columns

# Create a DataFrame from the reshaped input data with feature names
input_df = pd.DataFrame(input_data_reshaped, columns=feature_names)

# Standardize the input data
std_data = scaler.transform(input_df)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
